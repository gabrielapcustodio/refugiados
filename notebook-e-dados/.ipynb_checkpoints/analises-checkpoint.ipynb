# Número de refugiados com carteira assinada no Brasil cresce 150 vezes em 10 anos

```python
# importando dados. mude o local de acordo com o local onde os dados estao armazenados em seu google drive.
# link para a pasta com os arquivos: https://drive.google.com/drive/u/0/folders/1oL1wA4u1pco-zL_X3bC4bBnteJB4JRXQ
dados2011 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2011.csv", sep=";", encoding="latin1")
dados2012 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2012.csv", sep=";", encoding="latin1")
dados2013 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2013.csv", sep=";", encoding="latin1")
dados2014 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2014.csv", sep=";", encoding="latin1")
dados2015 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2015.csv", sep=";", encoding="latin1")
dados2016 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2016.csv", sep=";", encoding="latin1")
dados2017 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2017.csv", sep=";", encoding="latin1")
dados2018 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2018.csv", sep=";", encoding="latin1")
dados2019 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2019.csv", sep=";", encoding="latin1")
dados2020 = pd.read_csv("/content/drive/MyDrive/tcc-filipo-gabriela-mariana/dados/RAIS_CTPS_ESTOQUE_2020.csv", sep=";", encoding="latin1")
```

```python
# criando uma nova coluna com o ano
dados2011["ano"] = "2011"
dados2012["ano"] = "2012"
dados2013["ano"] = "2013"
dados2014["ano"] = "2014"
dados2015["ano"] = "2015"
dados2016["ano"] = "2016"
dados2017["ano"] = "2017"
dados2018["ano"] = "2018"
dados2019["ano"] = "2019"
dados2020["ano"] = "2020"
```

```python
# juntando as tabelas
dados = pd.concat([dados2011, dados2012, dados2013, dados2014, dados2015, 
                   dados2016, dados2017, dados2018, dados2019, dados2020])
```

```python
# decifrando alguns dados
dados["nivel_instrucao"] = dados["nivel_instrucao"].replace([1, 2, 3, 4, 5, 6, 7], ["Sem instrução ou ensino fundamental incompleto", "Ensino fundamental completo", "Ensino médio incompleto", "Ensino médio completo", "Ensino superior incompleto", "Ensino superior completo", "Pós-graduação"])
dados["uf"] = dados["uf"].replace([12, 27, 13, 16, 29, 23, 53, 32, 52, 21, 31, 50, 51, 15, 25, 26, 22, 41, 33, 24, 11, 14, 43, 42, 28, 35, 17], ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO'])
```

```python
# decifrando alguns dados - 2020
dados2020["nivel_instrucao"] = dados2020["nivel_instrucao"].replace([1, 2, 3, 4, 5, 6, 7], ["Sem instrução ou ensino fundamental incompleto", "Ensino fundamental completo", "Ensino médio incompleto", "Ensino médio completo", "Ensino superior incompleto", "Ensino superior completo", "Pós-graduação"])
dados2020["uf"] = dados2020["uf"].replace([12, 27, 13, 16, 29, 23, 53, 32, 52, 21, 31, 50, 51, 15, 25, 26, 22, 41, 33, 24, 11, 14, 43, 42, 28, 35, 17], ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO'])
```

```python
#ajustando numeros referentes a dinheiro
dados["vl_remun_media_sm"] = dados["vl_remun_media_sm"].str.replace(",", ".")
dados["vl_remun_media_sm"] = dados["vl_remun_media_sm"].astype(float)

dados2020["vl_remun_media_nom"] = dados2020["vl_remun_media_nom"].str.replace(",", ".")
dados2020["vl_remun_media_nom"] = dados2020["vl_remun_media_nom"].astype(float)

dados2020["vl_remun_media_sm"] = dados2020["vl_remun_media_sm"].str.replace(",", ".")
dados2020["vl_remun_media_sm"] = dados2020["vl_remun_media_sm"].astype(float)
```

```python
# Numero de refugiados ano a ano
total_ano = dados["ano"].value_counts()
total_ano = total_ano.to_frame()
total_ano = pd.DataFrame(total_ano.to_records())
total_ano.rename({"index": "ano", "ano": "total_ano"}, axis=1, inplace=True)
total_ano
```

```python
# Por status migratorio e por ano

status_ano = dados.groupby(by=["ano", "status_migratorio"])["ano"].agg("count")
status_ano = status_ano.to_frame()
status_ano.rename({"ano": "total"}, axis=1, inplace=True)
status_ano = pd.DataFrame(status_ano.to_records())
status_ano = pd.merge(status_ano, total_ano, on="ano", how="left")
status_ano['porc'] = (status_ano['total']/status_ano['total_ano'])*100
status_ano
```

```python
# Por pais e ano

pais_ano = dados.groupby(by=["ano", "pais"])["ano"].agg("count")
pais_ano = pais_ano.to_frame()
pais_ano.rename({"ano": "total"}, axis=1, inplace=True)
pais_ano = pd.DataFrame(pais_ano.to_records())
pais_ano = pd.merge(pais_ano, total_ano, on="ano", how="left")
pais_ano['porc'] = (pais_ano['total']/pais_ano['total_ano'])*100
pais_ano
```

```python
#migrantes por uf
uf_migrantes = dados2020["uf"].value_counts()
uf_migrantes = uf_migrantes.to_frame()
uf_migrantes = pd.DataFrame(uf_migrantes.to_records())
uf_migrantes.rename({"index": "uf", "uf": "total"}, axis=1, inplace=True)
uf_migrantes
```

```python
# salario medio geral de quem tem ensino superior completo
dados2020_superior = dados2020[dados2020["nivel_instrucao"] == "Ensino superior completo"]
dados2020_superior["vl_remun_media_sm"].mean()
```

```python
# salario medio geral
dados2020["vl_remun_media_sm"].mean()
```

```python
# salario medio geral de quem tem ensino superior completo por pais
dados2020_superior_pais = dados2020_superior.groupby(by=["pais"])["vl_remun_media_sm"].agg(["mean", "count"])
dados2020_superior_pais = pd.DataFrame(dados2020_superior_pais.to_records())
dados2020_superior_pais
```

```python
# salario mediano por continente de origem e grau de instrucao, em 2020
instrucao_cont_mediano = dados2020.groupby(by=["continente", "nivel_instrucao"])["vl_remun_media_sm"].agg(["median", "count"])
instrucao_cont_mediano = pd.DataFrame(instrucao_cont_mediano.to_records())
instrucao_cont_mediano
```

```python
# salario medio por continente de origem e grau de instrucao, em 2020
instrucao_cont_ano_medio = dados2020.groupby(by=["continente", "nivel_instrucao"])["vl_remun_media_sm"].agg(["mean", "count"])
instrucao_cont_ano_medio = pd.DataFrame(instrucao_cont_ano_medio.to_records())
instrucao_cont_ano_medio
```

```python
# salario mediano por pais de origem, em 2020
media_pais = dados2020.groupby(by=["pais", "ano", "continente"])["vl_remun_media_sm"].agg(["mean", "count"])
media_pais = pd.DataFrame(media_pais.to_records())
media_pais = media_pais[media_pais["count"] > 999]
media_pais
```
